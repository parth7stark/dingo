```python
# import bilby
import numpy as np
import pandas as pd
import corner
import matplotlib.pyplot as plt

from dingo.gw.result import Result

```

```python
# Load the importance sampling result
# result = Result(file_name = "../outdir_GW150914_new/result/GW150914_data0_1126259462-4_importance_sampling.hdf5")

result = Result(file_name = "../outdir_GW200219_094415_new/result/GW200219_094415_data0_1266140673-1_importance_sampling.hdf5")

samples = result.samples  # Dictionary of posterior samples
```

# Extract paramters of interest

```python
theta_jn = samples['theta_jn']  # Inclination angle in radians
median_theta_jn = np.median(theta_jn)
theta_jn_90_low, theta_jn_90_high = np.percentile(theta_jn, [5, 95])
print(f"Inclination angle: {np.degrees(median_theta_jn):.1f}° "
      f"({np.degrees(theta_jn_90_low):.1f}°–{np.degrees(theta_jn_90_high):.1f}°, 90% CI)")
```

```python
d_l = samples['luminosity_distance']  # Distance in megaparsecs (Mpc)
median_d_l = np.median(d_l)
d_l_90_low, d_l_90_high = np.percentile(d_l, [5, 95])
print(f"Luminosity distance: {median_d_l:.1f} Mpc "
      f"({d_l_90_low:.1f}–{d_l_90_high:.1f} Mpc, 90% CI)")
```

```python
m1 = samples['mass_1']  # Primary mass in solar masses
m2 = samples['mass_2']  # Secondary mass in solar masses

median_m1 = np.median(m1)
median_m2 = np.median(m2)
m1_90_low, m1_90_high = np.percentile(m1, [5, 95])
m2_90_low, m2_90_high = np.percentile(m2, [5, 95])

print(f"Primary mass: {median_m1:.1f} M☉ ({m1_90_low:.1f}–{m1_90_high:.1f} M☉, 90% CI)")
print(f"Secondary mass: {median_m2:.1f} M☉ ({m2_90_low:.1f}–{m2_90_high:.1f} M☉, 90% CI)")
```

```python
# Function to compute median and 90% credible interval
# Convert to DataFrame for easier handling
df = pd.DataFrame(samples)

def get_median_ci(data):
    median = np.median(data)
    low, high = np.percentile(data, [5, 95])
    return median, low, high

# 1. Inclination Angle (theta_jn)
theta_jn_median, theta_jn_low, theta_jn_high = get_median_ci(df['theta_jn'])
print(f"Inclination angle: {np.degrees(theta_jn_median):.1f}° "
      f"({np.degrees(theta_jn_low):.1f}°–{np.degrees(theta_jn_high):.1f}°, 90% CI)")

# 2. Luminosity Distance
dl_median, dl_low, dl_high = get_median_ci(df['luminosity_distance'])
print(f"Luminosity distance: {dl_median:.1f} Mpc "
      f"({dl_low:.1f}–{dl_high:.1f} Mpc, 90% CI)")

# 3. Sky Localization (RA, Dec)
ra_median, ra_low, ra_high = get_median_ci(df['ra'])
dec_median, dec_low, dec_high = get_median_ci(df['dec'])
print(f"RA: {np.degrees(ra_median):.1f}° ({np.degrees(ra_low):.1f}°–{np.degrees(ra_high):.1f}°, 90% CI)")
print(f"Dec: {np.degrees(dec_median):.1f}° ({np.degrees(dec_low):.1f}°–{np.degrees(dec_high):.1f}°, 90% CI)")

# 4. Masses
# Convert chirp mass and mass ratio to component masses
def chirp_mass_and_q_to_masses(chirp_mass, q):
    m1 = chirp_mass * (1 + q)**(1/5) * q**(-3/5)
    m2 = m1 * q
    return m1, m2

m1, m2 = chirp_mass_and_q_to_masses(df['chirp_mass'], df['mass_ratio'])
m1_median, m1_low, m1_high = get_median_ci(m1)
m2_median, m2_low, m2_high = get_median_ci(m2)
print(f"Primary mass: {m1_median:.1f} M☉ ({m1_low:.1f}–{m1_high:.1f} M☉, 90% CI)")
print(f"Secondary mass: {m2_median:.1f} M☉ ({m2_low:.1f}–{m2_high:.1f} M☉, 90% CI)")

# 5. Spins
# Compute effective spin
chi_eff = (m1 * df['a_1'] + m2 * df['a_2']) / (m1 + m2)
chi_eff_median, chi_eff_low, chi_eff_high = get_median_ci(chi_eff)
print(f"Effective spin: {chi_eff_median:.3f} ({chi_eff_low:.3f}–{chi_eff_high:.3f}, 90% CI)")
```
